# Neural E-commerce Data Exploration (EDA)

This notebook demonstrates how to load and analyze the demographic and sequential behavioral datasets exported from the platform.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")
```

## 1. Load Data

```python
# Load Demographic Data
df_users = pd.read_csv('../exports/demographic.csv')
print(f"Users loaded: {len(df_users)}")
display(df_users.head())
```

```python
# Load Sequential Behavioral Data
df_events = pd.read_parquet('../exports/sequential.parquet')
print(f"Events loaded: {len(df_events)}")
display(df_events.head())
```

## 2. Demographic Analysis

```python
# Age Group Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df_users, x='age_group', order=['18-24', '25-34', '35-44', '45+'], palette='viridis')
plt.title('Distribution of Users by Age Group')
plt.show()
```

## 3. Behavioral Analysis

```python
# Event Type Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df_events, y='event_type', order=df_events['event_type'].value_counts().index, palette='magma')
plt.title('Frequency of Behavioral Events')
plt.show()
```

```python
# Join Events with Demographics to analyze behavior by cohort
df_merged = df_events.merge(df_users[['customer_id', 'age_group', 'loyalty_tier']], on='customer_id', how='left')

# Example: Average time on page by Loyalty Tier
avg_time = df_merged[df_merged['event_type'] == 'view'].groupby('loyalty_tier')['time_on_page_sec'].mean().reset_index()
display(avg_time)
```
